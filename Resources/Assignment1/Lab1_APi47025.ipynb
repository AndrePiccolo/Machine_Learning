## e.Changing column labels

```python
#Removing unnecessary whitespaces from columns
carDataFrame.columns = carDataFrame.columns.str.strip() # Removing spaces from begin and end
carDataFrame.columns = carDataFrame.columns.str.replace(" ","") # Removing spaces between words
print(carDataFrame.info())
```

```python
# Set all columns to lowercase characters
carDataFrame.columns = carDataFrame.columns.str.lower()
print(carDataFrame.info())
```

## f.Dropping and filling data

```python
# Check null values per column
print(carDataFrame.isnull().sum())
```

```python
# Drop all rows where all data is missing
# Using how='all' to select row that all data is null
carDataFrame.dropna(how='all', inplace=True)
print(carDataFrame.isnull().sum())
```

```python
# Keep rows that have at least 9 non NaN values
carDataFrame.dropna(thresh=9, inplace=True)
print(carDataFrame.isnull().sum())
```

```python
# Fill the missing value in sellingprice with the values in auctionprice
carDataFrame.sellingprice.fillna(carDataFrame.auctionprice, inplace=True)

# Dropping the auctionprice because there are more null values on it than in sellingprice
carDataFrame.drop(columns={'auctionprice' }, axis=1, inplace=True)

# Change column name from sellingprice to price
carDataFrame.columns = carDataFrame.columns.str.replace("sellingprice","price")
print(carDataFrame.info())
```

```python
# Remove all null data from price column
carDataFrame.drop(carDataFrame.loc[carDataFrame['price'].isnull()].index, inplace=True)
print(carDataFrame.isnull().sum())
```

```python
# Remove rows that year/make and model are null
# Find record using loc with 'and' conditions to check both conditions together
carDataFrame.drop(carDataFrame.loc[carDataFrame['year/make'].isnull() & carDataFrame['model'].isnull()].index, inplace=True)
print(carDataFrame.isnull().sum())
```

```python
# Copy data transmission from columns that have the same year/make and model
# This operation is taken more time to execute, almost 3 minutes to finish this process
indexMissing = carDataFrame.loc[carDataFrame['transmission'].isnull()].index # return index that transmission is null

# Iterate for each index to get 'year/make' value to find a record with same 'year/make' and 'transmission' not null
for position in indexMissing:
    # Get a index list with just the first position 'head(1)'
    listTrasmissionResult = carDataFrame.loc[(carDataFrame['year/make'] == carDataFrame.loc[position, 'year/make'])&(carDataFrame['transmission'].notnull()),['transmission']].head(1).index.tolist()
    if(listTrasmissionResult): # Check if the list is not empty, this means that there are another 'year/make' register with 'transmission' value. Otherwise, just skip this value.
        carDataFrame.loc[position, 'transmission'] = carDataFrame.loc[listTrasmissionResult[0], 'transmission'] # Write value in null 'transmission' row

print(carDataFrame.isnull().sum())
```

```python
# Create the year and make column by splitting the year/make column using delimiter ‘/’
# 'expand=True' to create new columns
carDataFrame[['year','make']] = carDataFrame['year/make'].str.split('/', expand=True)
carDataFrame.head(1)
```

```python
# Drop unnecessary columns such as record, year/make, vin, state, seller, saledate
carDataFrame.drop(columns={'record','year/make','vin','state', 'seller', 'saledate'}, axis=1, inplace=True)
carDataFrame.head(1)
```

```python
# Rearrange the columns => year, make, model, trim, body, transmission, condition, odometer, color, interior, mmr, price
reorderColumns = ['year','make','model','trim','body','transmission','condition','odometer','color','interior','mmr','price']
carDataFrame = carDataFrame.reindex(columns=reorderColumns)
carDataFrame.head(1)
```

```python
# Drop all rows that still has nan/null values
carDataFrame.dropna(inplace=True)
print(carDataFrame.isnull().sum())
```

## g.Changing Columns Format and datatype

```python
# Format the column price such that you do not have the “$” sign and the ‘’,”
carDataFrame['price'] = carDataFrame['price'].str.replace("$","", regex='False')
carDataFrame['price'] = carDataFrame['price'].str.replace(",","", regex='False')
carDataFrame.head(2)
```

```python
# Set the datatype of price column into float64 and year into int32
carDataFrame.price = carDataFrame.price.astype('float64')
carDataFrame.year = carDataFrame.year.astype('int32')
print(carDataFrame.info())
```

```python
# Reduce the values of car body types
# Change any body type containing the word coupe or koup into coupe
carDataFrame.loc[carDataFrame.body.str.contains('(?i)coupe|koup'),'body'] = 'coupe'

# Find any record containing the word cab or crew and change the body type into truck
carDataFrame.loc[carDataFrame.body.str.contains('(?i)cab|crew'),'body'] = 'truck'

# Find any record containing the word sedan and change the body type into sedan
carDataFrame.loc[carDataFrame.body.str.contains('(?i)sedan'),'body'] = 'sedan'

# Find any record containing the word convertible and change the body type into convertible
carDataFrame.loc[carDataFrame.body.str.contains('(?i)convertible'),'body'] = 'convertible'

# Find any record containing the word wagon and change the body type into wagon
carDataFrame.loc[carDataFrame.body.str.contains('(?i)wagon'),'body'] = 'wagon'

# Find any record containing the word e-series van, transit van and any other van except the minivan and change the body type into van
carDataFrame.loc[carDataFrame.body.str.contains('(?i)e-series|transit|ram|promaster'),'body'] = 'van'
carDataFrame.body.value_counts()

# Your dataset should only contain the following body type: coupe, sedan, truck, minivan, convertible, wagon, hatchback, and van
# There is no matching for SUV, I`m not sure if can change SUV for any available body type or just let SUV in this list
```

```python
# Change the color value of the records with color “-“ and any color whose value_counts() is less than the value_counts() of “-“ to other.
lineOptionColor = (carDataFrame.color.values == '—').sum() # Get total of '—' to use in if condition 
for colorName in carDataFrame.color.unique(): # Get the colors from unique() method
    if((carDataFrame.color.values == colorName).sum() <= lineOptionColor): # if actual color ir smaller than '—', so change 'color' value to 'other' 
        carDataFrame.loc[carDataFrame.color.str.contains(colorName),'color'] = 'other'
print(carDataFrame.color.value_counts())
```

```python
# Change the interior value of the records with color “-“ and any interior color whose value_counts() is less than the value_counts() of “-“ to other.
# Using the same logic from previous operation
lineOptionInterior = (carDataFrame.interior.values == '—').sum()
for interiorName in carDataFrame.interior.unique():
    if((carDataFrame.interior.values == interiorName).sum() <= lineOptionInterior):
        carDataFrame.loc[carDataFrame.interior.str.contains(interiorName),'interior'] = 'other'
print(carDataFrame.interior.value_counts())
```
